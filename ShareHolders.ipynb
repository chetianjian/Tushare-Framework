```python
import tushare as ts
import numpy as np
import pandas as pd
import time
```

```python
pro = ts.pro_api()  # Token needed for access.
```

```python
stocks_info = pro.query("stock_basic", exchange="", list_status="L", 
                        fields="ts_code,symbol,name,area,industry,list_date")
code_lst = stocks_info["ts_code"]
```

```python
data_df = pd.DataFrame(columns=["ts_code", "ann_date", "end_date", "holder_name",
                                "hold_amount", "hold_ratio"])
temp_df = data_df.copy()
```

```python
count, total = 0, len(code_lst)
empty_code = []

for code in code_lst:
    count += 1
    print(f"Now saving {count}/{total}...")
    df = pro.query("top10_holders", ts_code=code, start_date="20220101", end_date="20220430")
    if not df.empty:
        temp_df = pd.concat([temp_df, df])
    else:
        empty_code.append(code)
    if count % 5 == 0:
        data_df = pd.concat([data_df, temp_df])
        temp_df = pd.DataFrame(columns=["ts_code", "ann_date", "end_date", 
                                        "holder_name","hold_amount", "hold_ratio"])
        data_df.to_csv("C:\\Users\\tianj\\Desktop\\ShareHolders.csv", encoding="utf_8_sig")
    if count % 10 == 0:
        print(f"The last stock we get: {code}")
        print("Wait for 65 seconds...")
        time.sleep(65)
```
